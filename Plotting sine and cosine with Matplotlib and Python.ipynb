# Trignometric Functions Sin(x), cos(x)

```python
# Matplotlib is a plotting library that can produce line plots, bar graphs, histograms and many other types of plots using Python. 
# Matplotlib is not included in the standard library. 
# install matplotlib and numpy with pip on the command line. 
# pip install matplotlib
# pip install numpy
# If you are using the Anaconda distribution of Python, no need to install any additional packages to use matplotlib.

import matplotlib.pyplot as plt
import numpy as np
# if using a jupyter notebook
# %matplotlib inline

# build a set of x values from zero to 4Ï€ in increments of 0.1 radians.
# The x-values are stored in a numpy array.

x = np.arange(0, 4*np.pi, 0.1) # start,stop,step
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
# build a plot which shows two trig functions, sine and cosine
x = np.arange(0, 4*np.pi, 0.1) # start,stop,step
y = np.sin(x)
z = np.cos(x) # numpy array z
plt.plot(x,y,x,z)
plt.show()
```

```python
# Example 2: with labels, title, legend

x = np.arange(0,4*np.pi-1,0.1)   # start,stop,step size
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y,x,z)
plt.xlabel('x values from 0 to 4pi')
plt.ylabel('sin(x) and cos(x)')
plt.title('Plot of sin and cos from 0 to 4pi')
plt.legend(['sin(x)','cos(x)']) # legend entries as seperate strings in a list
plt.show()
```

```python
# Example 3: y=sin(x) for x between âˆ’Ï€ and Ï€

import matplotlib.pyplot as plt
import numpy as np

# Syntax: numpy.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)
# 100 linearly spaced numbers
x = np.linspace(-np.pi, np.pi, 100)  # start,stop,number of samples to generate (default 50)
y = np.sin(x)

# setting the axes at the centre
# Syntax: matplotlib.pyplot.figure(num=None, figsize=None, dpi=None, facecolor=None, edgecolor=None, frameon=True, 
#                                  FigureClass=<class 'matplotlib.figure.Figure'>, clear=False, **kwargs)
# Create a new figure, or activate an existing figure.
fig = plt.figure()

# pyplot.subplots creates a figure and a grid of subplots with a single call
# Syntax: subplot(nrows, ncols, index, **kwargs)
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# plot the function
# Syntax: matplotlib.pyplot.plot(*args, scalex=True, scaley=True, data=None, **kwargs)
plt.plot(x,y, 'b-') # Ex: b+, ro

# show the plot
plt.show()
```

```python
# Example 4: 
import numpy as np
import matplotlib.pyplot as plot

# Get x values of the sine wave
time = np.arange(0, 10, 0.1)

# Amplitude of the sine wave is sine of a variable like time
amplitude = np.sin(time)

# Plot a sine wave using time and amplitude obtained for the sine wave
plot.plot(time, amplitude)
plot.title('Sine wave')
plot.xlabel('Time')
plot.ylabel('Amplitude = sin(time)')
plot.grid(True, which='both')

# Syntax: matplotlib.pyplot.axhline(y=0, xmin=0, xmax=1, **kwargs)
# Add a horizontal line across the axis.
plot.axhline(y=0, color='k') # k black
plot.show()
```

# Exponential Graph using Python

```python
# Ex 1: **

base = 2
exponent = 4
print(base ** exponent)
```

```python
# Ex 2: pow(base, exponent) buit in function in python

print(pow(base, exponent))
```

```python
# Ex 3: exp() returns exponential of x.
# It allows user to calculate the exponential value with the set e.
# e = 2.71828183

import math
print(math.exp(exponent))
```

```python
# Ex 4:
# Syntax: numpy.exp(array, out = None, where = True, casting = â€˜same_kindâ€™, order = â€˜Kâ€™, dtype = None)

import numpy as np

#create a list
l=[1,2,3,4,5]
l1 = np.array([1,2,3,4,5])
print(l)
print(np.exp(l))
print(np.exp(l1))
```

```python
# Ex 5: numpy.exp() with matplotlib

import numpy as np
import matplotlib.pyplot as plt
x= np.array([1,2,3,4,5])
y=np.exp(x)
plt.plot(x,y)
plt.show()
```

```python
# Ex 6: plot two graphs together

import numpy as np
import matplotlib.pyplot as plt
x= np.array([1,3,5])
y=[2,4,6]
z=np.exp(x)
w=np.exp(y)
plt.plot(x,z,color="blue",marker='*')
plt.plot(y,w,color="red",marker='o')
plt.xlabel("X Axis--------->")
plt.ylabel("Y Axis--------->")
plt.show()
```

```python
x = np.arange(-3.14, 3*3.14)
a = 2 * x
b = x * 3
y = np.abs(a * x + b)
plt.plot(x,y)
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Absolute function')
plt.grid(True)
plt.show()
```

Logarithm of a function

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0, 3.14, 0.01)
a = 4
b = 5
y = np.log(a * x + b)
plt.plot(x, y)
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Logarithm')
plt.grid('True')
plt.show()
```

Plotting Graph of 1/ax+b

```python
x = np.arange(0, 3*np.pi, 0.01)
a = 4
b = 5
y = 1/(a*x+b)
plt.plot(x, y)
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('1/ax+b')
plt.grid(True)
plt.show()
```

Parabola

```python
# Ex:1
# create 1000 equally spaced points between -10 and 10
x = np.linspace(-10, 10, 1000)

# calculate the y value for each element of the x vector
y = x**2 + 2*x + 2  

fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
# Ex 1: another way
a=[]
b=[]

for x in range(-50,50,1):
    y=x**2+2*x+2
    a.append(x)
    b.append(y)

fig= plt.figure()
axes=fig.add_subplot(111)
axes.plot(a,b)
plt.show()
```

```python
#Ex : 2
x = np.linspace(-100, 100)
a = 1
b = 2
c = 36
y = a*x**2 + b*x + c
plt.plot(x, y)
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Parabola')
plt.grid(True)
plt.show()
```

Hyperbola

```python
x = np.linspace(-100, 100)
a = 1
b = 2
y = np.sqrt((x/a)**2-1 * b**2)
plt.plot(x, y)
plt.plot(x, -y)
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Hyperbola')
plt.grid(True)
plt.show()
```

Ellipse

```python
x = np.linspace(-2, 2)
a = 2
b = 4
y = np.sqrt(1-(x/a)**2)*b**2
plt.plot(x, y)
plt.plot(x, -y)
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Hyperbola')
plt.grid(True)
plt.show()
```

Surface of Parabola

```python
x = np.linspace(-4, 4)
y = np.linspace(-4, 4)
x,y = np.meshgrid(x, y)
z = y**2+4*x
plt.figure()
levels = [-2,-1,0,1]
plt.contourf(x,y,z,levels)
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Surface of Parabola')
plt.grid(True)
plt.show()
```

Surface of Hyperbola

```python
x = np.linspace(-3, 3)
y = np.linspace(-3, 3)
x,y = np.meshgrid(x, y)
z = 1+x**2-y**2
plt.figure()
levels = [-2,-1,0,1]
plt.contourf(x,y,z,levels)
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Surface of Hyperbola')
plt.grid(True)
plt.show()
```

Surface of Ellipse

```python
x = np.linspace(-3, 3)
y = np.linspace(-3, 3)
x,y = np.meshgrid(x, y)
z = 1-x**2-y**2
plt.figure()
levels = [-2,-1,0,1]
plt.contourf(x,y,z,levels)
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Surface of Ellipse')
plt.grid(True)
plt.show()
```
